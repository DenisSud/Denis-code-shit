```python
import numpy as np 
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn import metrics
import random
```

```python
training_set_df = pd.read_csv("train.csv")
training_set_df.describe()
```

```python
training_set_df.dropna(subset=['Embarked'], inplace = True)
training_set_df.shape
```

```python
training_set_df = training_set_df.drop(['PassengerId','Name', 'Ticket', 'Cabin'], axis=1)
training_set_df
```

```python
training_set_df['Age'].fillna(training_set_df['Age'].median(), inplace = True)
# training_set_df['Sex'].fillna(training_set_df['Sex'].mod(), inplace = True)
training_set_df
```

```python
training_set_df['Sex'] = (training_set_df['Sex'].map( {'female': 1, 'male': 0} )).astype(int)
training_set_df['Embarked'] = training_set_df['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
```

```python
X = training_set_df.drop('Survived', axis=1)
y = training_set_df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2)
model = KNeighborsClassifier(5)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
sensitivity = (model.predict(X_test): y_test).sensetivsty
print(sensitivity)
# true = 0
# false = 0   
# true_guess = 0
# false_guess = 0
# for n in y_test:
#     if n == 1:
#         true += 1
#     else:
#         false += 1
# for m in range(len(y_test)-1):
#     if predictions[m] == y_test[m] and y_test[m] == 1:
#         true_guess += 1
#     elif predictions[m] == y_test[m] and y_test[m] == 0:
#         false_guess += 1
#     sensitivity = true/true_guess
#     specificity = false/false_guess
# print(sensitivity)
# print(specificity)
```
